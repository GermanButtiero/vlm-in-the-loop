```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('white') 
ax.set_facecolor('white')  

# Function to add a trace with numbered markers
def add_numbered_trace(x_values, y_values, color, name):
    # Plot line with circle markers
    ax.plot(x_values, y_values, '-o', color=color, markersize=1, label=name,linewidth=0.8) 
    
    # Add text labels (numbers) next to each point
    for i, (x, y) in enumerate(zip(x_values, y_values)):
        ax.annotate(str(i), 
                    xy=(x, y),
                    xytext=(-10, 5),  # Offset text by 5 points
                    textcoords='offset points',
                    color=color)
        
def add_numbered_trace_human(x_values, y_values, color, name, style_option=1):
    # Define style options
    styles = {
         1: '-o',    # Solid line with circles
        2: '-+',    # Solid line with plus signs
        3: '-^',    # Solid line with triangles
        4: '-s',    # Solid line with squares
        5: '-*'     # Solid line with stars
    }
    
    # Get the selected style or default to option 1
    style = styles.get(style_option, '-o')
    
    # Plot with the selected style
    ax.plot(x_values, y_values, style, color=color, markersize=3, label=name, linewidth=0.8)
    
    # Add text labels (numbers) next to each point
    for i, (x, y) in enumerate(zip(x_values, y_values)):
        ax.annotate(str(i), 
                    xy=(x, y),
                    xytext=(5, 5),  # Offset text by 5 points
                    textcoords='offset points',
                    color=color)
# Add each dataset with numbered markers
add_numbered_trace(
    vlm_0_2_x_axis * len(vlm_0_2), 
    vlm_0_2, 
    'red', 
    'VLM-Active 20%'
)
add_numbered_trace_human(
    human_0_2_x_axis,
    human_0_2, 
    "black",
    'Human-Active 20%',1
)

add_numbered_trace(
    vlm_0_4_x_axis * len(vlm_0_4), 
    vlm_0_4, 
    'blue', 
    'VLM-Active 40%'
)

add_numbered_trace_human(
    human_0_4_x_axis,
    human_0_4, 
    "black",
    'Human-Active 40%',2
)

add_numbered_trace(
    vlm_0_6_x_axis * len(vlm_0_6), 
    vlm_0_6, 
    'green', 
    'VLM-Active 60%'
)
add_numbered_trace_human(
    human_0_6_x_axis,
    human_0_6, 
    "black",
    'Human-Active 60%',3
)
add_numbered_trace(
    vlm_0_8_x_axis * len(vlm_0_8), 
    vlm_0_8, 
    'orange', 
    'VLM-Active 80%'
)
add_numbered_trace_human(
    human_0_8_x_axis,
    human_0_8, 
    "black",
    'Human-Active 80%',4
)
add_numbered_trace_human(
    human_1_x_axis ,
    human_1, 
    "black",
    "MaskRCNN fully supervised",5
)

# Add grid for better readability
ax.grid(True, alpha=0.7, color='lightgray')

# Set labels and title
ax.set_xlabel('Annotation Time (hs)')
ax.set_ylabel('Mean AP')
ax.set_title('Performance vs Annotation Time')

# Adjust layout
plt.tight_layout()

# Add legend
plt.legend(fontsize=8, frameon=True, framealpha=0.8, loc='lower right', ncol=2, markerscale=0.8)


# Show the figure
plt.show()
```
